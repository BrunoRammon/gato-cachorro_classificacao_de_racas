# Functions

```python
def plot_avg_accuracies(filename,xlabel):
    df_knn = pd.read_csv(filename)
    df_knn.set_index(['dataset','train/test_type'],inplace=True)
    knn_means = df_knn.describe().loc['mean',:]
    ax = sns.lineplot(y=knn_means.values, x=knn_means.index)
    ax.set_ylabel('acc')
    ax.set_xlabel(xlabel)
    # image_name=filename.split('.')[0]
    # plt.savefig(f'{image_name}.pdf')
```

# Knn

```python
file_name = this_absolute_path+'knn_accuracies.csv'
plot_avg_accuracies(file_name,'Número de vizinhos mais próximos')
```

```python
file_name = this_absolute_path+'knn_accuracies_new.csv'
plot_avg_accuracies(file_name,'Número de vizinhos')
```

# Bagging

```python
estimators = ['knn','dt','mlp','nb']
for estimator in estimators:
    file_name = this_absolute_path+f'bagging_{estimator}_accuracies.csv'
    plot_avg_accuracies(file_name,'Número de estimadores')
labels=[f'{estimator}' for estimator in estimators]
plt.legend(labels=labels);
```

# Random forest

```python
import regex as re
df = pd.read_csv(this_absolute_path+f'random_forest_accuracies.csv')
df.set_index(['dataset','train/test_type'],inplace=True)
df.columns = ['/'.join(re.findall(r"[a-z].*=([\d].+),[a-z].*=([\w].+)",col)[0])for col in df.columns]
df_means = df.describe().loc['mean',:]
ax = sns.lineplot(y=df_means.values, x=df_means.index)
ax.set_ylabel('acc')
ax.set_xlabel('Número de estimadores/Critério de importância de atributo')
plt.legend(labels=['accuracy']);
```
